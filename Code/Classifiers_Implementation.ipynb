```python
import pandas as pd
import numpy as np
df1 = pd.read_csv('/content/tru1.csv')
```

```python
# Remove duplicate rows
cleaned_df = df1.drop_duplicates()
# Replace missing values in 'column' with 0
# Replace NaN values with zeros
df1 = df1.fillna(0)
```

**Manual Features Selection**

```python
# All features together
x = df1.drop(columns="class label").values
y = df1["class label"].values
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import pandas as pd

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)


models = {
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(),
    "Logistic Regression": LogisticRegression(),
     "KNeighborsClassifier": KNeighborsClassifier(),
    "Decision Trees": DecisionTreeClassifier()


}

results = {}
model_results = {}
accur = {}
accur_re = {}
phishing_y_test = []
regular_y_test = []
phishing_y_pred = []
regular_y_pred = []

for model_name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)

    # Predict on the test set
    y_pred = model.predict(X_test)

    # Calculate accuracy, precision, and recall
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1_scor = f1_score(y_test, y_pred)
    model_results[model_name] = {"Accuracy": accuracy, "Precision": precision, "Recall": recall,"f1_scor":f1_scor}
    accur[model_name] = accuracy
    for index, prediction in enumerate(y_pred):
          email = y_test[index]
          if email == 1:
             phishing_y_pred.append(prediction)
          else:
            regular_y_pred.append(prediction)
    for index, prediction in enumerate(y_test):
          email = y_test[index]
          if email == 1:
               phishing_y_test.append(prediction)
          else:
              regular_y_test.append(prediction)
    regular_accuracy = accuracy_score(regular_y_test, regular_y_pred)
    regular_precision = precision_score(regular_y_test, regular_y_pred, pos_label=0)
    regular_recall = recall_score(regular_y_test, regular_y_pred, pos_label=0)
    regular_f1 = f1_score(regular_y_test, regular_y_pred, pos_label=0)

    # Calculate metrics for phishing emails
    phishing_accuracy = accuracy_score(phishing_y_test, phishing_y_pred)
    phishing_precision = precision_score(phishing_y_test, phishing_y_pred)
    phishing_recall = recall_score(phishing_y_test, phishing_y_pred)
    phishing_f1 = f1_score(phishing_y_test, phishing_y_pred)

    # Create labels for the metrics
    metrics = ['Regular', 'Phishing']

    # Create values for the metrics
    accuracy_values = [regular_accuracy, phishing_accuracy]
    precision_values = [regular_precision, phishing_precision]
    recall_values = [regular_recall, phishing_recall]
    f1_values = [regular_f1, phishing_f1]

    # Set the positions of the bars on the x-axis
    ind = np.arange(len(metrics))

    # Set the width of each bar
    width = 0.2

    # Plot the accuracy
    plt.bar(ind, accuracy_values, width, label='Accuracy', color='blue')
    # Plot the precision
    plt.bar(ind + width, precision_values, width, label='Precision', color='orange')
    # Plot the recall
    plt.bar(ind + 2 * width, recall_values, width, label='Recall', color='green')
    # Plot the F1-score
    plt.bar(ind + 3 * width, f1_values, width, label='F1-score', color='red')

    # Set the x-axis labels and title
    plt.xticks(ind + 1.5 * width, metrics)
    plt.xlabel('Email Type')
    plt.ylabel('Score')
    plt.title(f"{model_name} Performance Metrics")
    # Set the y-axis limits
    plt.ylim(0.8, 1.0)
    # Adjust the legend position outside the plot
    plt.legend(loc='upper left', bbox_to_anchor=(-0.095, 1.26))

    # Show the plot
    plt.show()

results["features 1"] = model_results
accur_re["features 1"] = accur
# Create a DataFrame from the results
df_results = pd.DataFrame.from_dict(results, orient='columns')
df1_results = pd.DataFrame.from_dict(accur_re, orient='columns')
# Plot the results
ax = df1_results.plot(kind="bar", figsize=(10, 6))
ax.set_ylim([0.8, 1.0])  # Set y-axis limits
plt.xlabel("Models")
plt.ylabel("Score")
plt.title("Comparison of Model Performance")
plt.legend(title="Datasets")
plt.xticks(rotation=0)

# Create a table for precision and recall comparison
df_precision_recall = pd.DataFrame(model_results).T
df_precision_recall = df_precision_recall[["Accuracy", "Precision", "Recall","f1_scor"]]

# Print precision and recall comparison table
print("metrices Comparison:")
print(df_precision_recall)
plt.show()
```
